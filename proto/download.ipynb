```python
import osmnx as ox

G = ox.graph_from_place('Madrid, Spain', network_type="drive_service")
ox.plot_graph(G)
```

```python
G_project = ox.project_graph(G)
G_consolidated = ox.consolidate_intersections(G_project, tolerance=10, rebuild_graph=True, dead_ends=True)

ox.plot_graph(G_consolidated)
```

```python
ox.io.save_graph_geopackage(G, filepath="madrid.gpkg")
```

```python
ox.io.save_graphml(G, filepath="madrid.gml")
```

```python
G_reloaded = ox.io.load_graphml(filepath="madrid.gml")
ox.plot_graph(G_reloaded)
```
